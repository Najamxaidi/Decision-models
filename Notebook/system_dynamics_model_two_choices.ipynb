```python
# %pylab inline
%matplotlib inline

import sys
import numpy as np
sys.path.append('../')

import system_dynamics_using_sdeint as systemd
```

Lets create a model with almost no noise initially. There are two choices avalible 0 and 1. Utility of choice 0 is less than utility of choice 1 but the experience of choice 0 is greater than the experience of choice 1. Furthermore there is no switching among the utilities.

```python
sysd = systemd.SystemDynamicsWithSdeint(number_of_agents= 100, k = 0.5, alpha = 2, utility_of_choices= [10, 20],
                          initial_experiences= [100, 10], discount_rate=0.01, sd = 0, rotation_time = 2000, 
                                        flag = False)

time_vector=np.linspace(0, 1000, 100)
sysd.solve(time_vector)
```

We can observe that all the agents are prefering choice 0 although choice 1 has a higher utility
Lets bring in some noise

```python
sysd = systemd.SystemDynamicsWithSdeint(number_of_agents= 100, k = 0.5, alpha = 2, utility_of_choices= [10, 20],
                          initial_experiences= [100, 10], discount_rate=0.01, sd = 100, rotation_time = 2000, 
                                        flag = False)

time_vector=np.linspace(0, 1000, 100)
sysd.solve(time_vector)
```

The experience have started to change indicating the agents have started to explore the other option as well Lets increase the noise further
Increasing noise further

```python
sysd = systemd.SystemDynamicsWithSdeint(number_of_agents= 100, k = 0.5, alpha = 2, utility_of_choices= [10, 20],
                          initial_experiences= [100, 10], discount_rate=0.01, sd = 2000, rotation_time = 2000, 
                                        flag = False)

time_vector=np.linspace(0, 1000, 100)
sysd.solve(time_vector)
```

In the presence of noise agents were able to shift to an option of higher utilty.
