## System dynamices model
### No noise

In the scnerios below phi would be varied in each case to move the system between three possible regions og the grpah

### Explotation of the multiple choices

To get (phi * q)/2 in this area phi = 8 has been used it gives a value (phi * q)/2 = 2

```python
sysd = systemd.SystemDynamics(number_of_agents=8, k=1, alpha=2,
                          utility_of_choices=[0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 
                                              0.5, 0.5, 0.5, 0.5, 0.5, 0.5,
                                              0.5],
                          initial_experiences=[0.10, 0.12, 0.13, 0.14, 0.15, 0.16, 0.17, 
                                               0.18, 0.19, 0.20, 0.21, 0.22, 0.23,
                                               0.24, 0.25, 0.10],
                          discount_rate=1)

sysd.solve()
```

An equal exploitation of multiple sources can be observed. PLEASE LOOK AT THE AREA UNDER THE CURVE THAT INDICATES EQUAL EXPLOTATION 

### Explotitation of a single or of a multiple source 

To get (phi * q)/2 in this area phi = 40 has been used it gives a value (phi * q)/2 = 10

```python
sysd = systemd.SystemDynamics(number_of_agents=10, k=1, alpha=2,
                          utility_of_choices=[0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 
                                              0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5,
                                              0.5],
                          initial_experiences=[0.10, 0.12, 0.13, 0.14, 0.15, 0.16, 0.17, 
                                               0.18, 0.19, 0.20, 0.21, 0.22, 0.23,
                                               0.24, 0.25, 0.10],
                          discount_rate=1)

sysd.solve()
```

An explotation of multiple resources can be observed
