```python
# %pylab inline
%matplotlib inline

import sys
import numpy as np
sys.path.append('../')

import system_dynamics_using_sdeint as systemd
import stat_gen_sys_dynamic as sgsd
```

# Noise makes the system follow changes in the environment

```python
#No noise 
# observe that system dont follow the changes in the environment

sysd = systemd.SystemDynamicsWithSdeint(number_of_agents=10,
                                    k=1,
                                    alpha=2,
                                    utility_of_choices=[0.25,0.50,0.75,1],
                                    initial_experiences=[1, 0.75, 0.50, 0.25],
                                    discount_rate=0.99,
                                    sd=0,
                                    rotation_time=200,
                                    flag=True)

time_vector=np.linspace(0, 1000, 500)
sysd.solve(time_vector)
sysd.return_area(time_vector)
```

```python
#increasing noise 
# observe that system still dont follow the changes in the environment

sysd = systemd.SystemDynamicsWithSdeint(number_of_agents=10,
                                    k=1,
                                    alpha=2,
                                    utility_of_choices=[0.25,0.50,0.75,1],
                                    initial_experiences=[1, 0.75, 0.50, 0.25],
                                    discount_rate=0.99,
                                    sd=0.5,
                                    rotation_time=200,
                                    flag=True)

time_vector=np.linspace(0, 1000, 500)
sysd.solve(time_vector)
sysd.return_area(time_vector)
```

```python
#increasing noise 
# observe that system starts to follow changes in the environment

sysd = systemd.SystemDynamicsWithSdeint(number_of_agents=10,
                                    k=1,
                                    alpha=2,
                                    utility_of_choices=[0.25,0.50,0.75,1],
                                    initial_experiences=[1, 0.75, 0.50, 0.25],
                                    discount_rate=0.99,
                                    sd=1.3,
                                    rotation_time=200,
                                    flag=True)

time_vector=np.linspace(0, 1000, 500)
sysd.solve(time_vector)
sysd.return_area(time_vector)
```

```python
#increasing noise 
# observe that system response is getting too much random

sysd = systemd.SystemDynamicsWithSdeint(number_of_agents=10,
                                    k=1,
                                    alpha=2,
                                    utility_of_choices=[0.25,0.50,0.75,1],
                                    initial_experiences=[1, 0.75, 0.50, 0.25],
                                    discount_rate=0.99,
                                    sd=2,
                                    rotation_time=200,
                                    flag=True)

time_vector=np.linspace(0, 1000, 500)
sysd.solve(time_vector)
sysd.return_area(time_vector)
```

```python
#increasing noise 
# observe that with too much noise no useful information is transferred

sysd = systemd.SystemDynamicsWithSdeint(number_of_agents=10,
                                    k=1,
                                    alpha=2,
                                    utility_of_choices=[0.25,0.50,0.75,1],
                                    initial_experiences=[1, 0.75, 0.50, 0.25],
                                    discount_rate=0.99,
                                    sd=4,
                                    rotation_time=200,
                                    flag=True)

time_vector=np.linspace(0, 1000, 500)
sysd.solve(time_vector)
sysd.return_area(time_vector)
```
