This paper replicate the results as presented by S. C. NICOLIS AND J. L. DENEUBOURG In their paper Emerging Patterns and Food Recruitment in Ants: an Analytical Study

One trail will be different where the rest of s-1 trails are identical. Three type of experiences will be there. One for the different source while other two for the s-1 dources. 

Observe in all cases that richer source always take over

```python
# lets setup the system dynamic model first

# %pylab inline
%matplotlib inline

import sys
import numpy as np
sys.path.append('../')

import system_dynamics_using_sdeint as systemd
```

#  Five choices scnerio
## System dynamices model
### No noise

```python
#parameters are decided based upon the discussion in paper --  see page 587 figure 6

sysd = systemd.SystemDynamicsWithSdeint(number_of_agents= 10, k = 0.5, alpha = 2, 
                                        utility_of_choices= [1, 0.9, 0.9, 0.9, 0.9],
                                        initial_experiences= [0.08, 0.11, 0.12, 0.10, 0.09], 
                                        discount_rate=0.01, 
                                        sd = 0, 
                                        rotation_time = 2000, 
                                        flag = False)

time_vector=np.linspace(0, 50, 25)
sysd.solve(time_vector)
```

We observe that when one choice is better than the others, this option always takes over

Lets now do it for the scneario when one choice is worst than all others

```python
#parameters are decided based upon the discussion in paper --  see page 587 figure 6

sysd = systemd.SystemDynamicsWithSdeint(number_of_agents= 10, k = 0.5, alpha = 2, 
                                        utility_of_choices= [0.9, 1, 1, 1, 1],
                                        initial_experiences= [0.08, 0.11, 0.12, 0.10, 0.09], 
                                        discount_rate=0.01, 
                                        sd = 0, 
                                        rotation_time = 2000, 
                                        flag = False)

time_vector=np.linspace(0, 150, 75)
sysd.solve(time_vector)
```

Observe the three oncentrations red + blue + rest of colors
