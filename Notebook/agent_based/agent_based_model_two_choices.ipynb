```python
import agent_based_model_for_any_number_of_choices as agentm
```

Lets create a model with almost no noise initially. There are two choices avalible 0 and 1. Utility of choice 0 is less than utility of choice 1 but the experience of choice 0 is greater than the experience of choice 1. Furthermore there is no switching among the utilities.

```python
steps = 200    # we want the model to run for 200 steps
rotation_step = 100 # utility should be inverted at this step
flag = False # utility inversion is off
```

```python
d = agentm.Agent_Based_Decision_Model(number_of_agents= 100, k = 0.5, alpha = 2,
                                   utility_of_choices= [10, 20], initial_experiences=[100, 10],
                                   discount_rate=0.01, noise_standard_deviation=0.1)

d.run(steps,rotation_step, flag)
```

We can observe that all the agents are prefering choice 0 although choice 1 has a higher utility

Lets bring in some noise

```python
d = agentm.Agent_Based_Decision_Model(number_of_agents= 100, k = 0.5, alpha = 2,
                                   utility_of_choices= [10, 20], initial_experiences=[100, 10],
                                   discount_rate=0.01, noise_standard_deviation=100)

d.run(steps,rotation_step, flag)
```

The experience have started to change indicating the agents have started to explore the other option as well
Lets increase the noise further

Increasing noise further

```python
d = agentm.Agent_Based_Decision_Model(number_of_agents= 100, k = 0.5, alpha = 2,
                                   utility_of_choices= [10, 20], initial_experiences=[100, 10],
                                   discount_rate=0.01, noise_standard_deviation=200)

d.run(steps,rotation_step, flag)
```

In the presence of noise agents were able to shift to an option of higher utilty.

# Switching utilities

```python
# lets turn on switching
flag = True # utility inversion is off
```

```python
# system with little noise 
d = agentm.Agent_Based_Decision_Model(number_of_agents= 100, k = 0.5, alpha = 2,
                                   utility_of_choices= [10, 20], initial_experiences=[100, 10],
                                   discount_rate=0.01, noise_standard_deviation=0.1)

d.run(steps,rotation_step, flag)
```

Observe that agents remain stucked in the wrong choice

```python
#Lets run it for the scnerio from above where switching occured

d = agentm.Agent_Based_Decision_Model(number_of_agents= 100, k = 0.5, alpha = 2,
                                   utility_of_choices= [10, 20], initial_experiences=[100, 10],
                                   discount_rate=0.01, noise_standard_deviation=200)

d.run(steps,rotation_step, flag)
```

We can see two switches indicating that noise helped the agents in making a better decision

Lets see what happens in the scnerio where they have already taken a better decision because of initial higher experiences but the utility is inverted and there is not much noise in the system

```python
# Small difference in the initial experiences - no noise
d = agentm.Agent_Based_Decision_Model(number_of_agents= 100, k = 0.5, alpha = 2,
                                   utility_of_choices= [10, 20], initial_experiences=[100, 75],
                                   discount_rate=0.01, noise_standard_deviation=0.1)

d.run(steps,rotation_step, flag)
```

Though they initially have taken the correct decision but the agents were not able to switch back to the better choice once the utilities have been swaped. We will now bring noise in this system and see what happens

```python
# Small difference in the initial experiences
d = agentm.Agent_Based_Decision_Model(number_of_agents= 100, k = 0.5, alpha = 2,
                                   utility_of_choices= [10, 20], initial_experiences=[100, 75],
                                   discount_rate=0.01, noise_standard_deviation=150)

d.run(steps,rotation_step, flag)
```

Noise was helpful in this scnerio
