## System dynamics model

```python
# lets setup the system dynamic model first

# %pylab inline
%matplotlib inline

import sys
import numpy as np
sys.path.append('../')

import system_dynamics_using_sdeint as systemd
```

```python
sysd = systemd.SystemDynamicsWithSdeint(number_of_agents= 10, k = 0.5, alpha = 2, 
                                        utility_of_choices= [1, 0.9, 0.9, 0.9, 0.9],
                                        initial_experiences= [0.08, 0.11, 0.12, 0.10, 0.09], 
                                        discount_rate=0.01, 
                                        sd = 0, 
                                        rotation_time = 2000, 
                                        flag = False)

time_vector=np.linspace(0, 50, 25)
sysd.solve(time_vector)
```
