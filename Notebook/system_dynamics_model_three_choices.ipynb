Lets now extend system dynamics model for three choices

# Simple scenario -- no rotation

```python
# %pylab inline
%matplotlib inline

import sys
import numpy as np
sys.path.append('../')

import system_dynamics_using_sdeint as systemd
```

```python
#simple case - no noise - large difference in experiences
sysd = systemd.SystemDynamicsWithSdeint(number_of_agents= 100, k = 0.5, alpha = 2, utility_of_choices= [10, 100, 20],
                          initial_experiences= [200, 2, 50], discount_rate=0.01, sd = 0, rotation_time = 2000, 
                                        flag = False)

time_vector=np.linspace(0, 1000, 100)
sysd.solve(time_vector)
```

As expected the choice with the higher initial experience is chosen

```python
#simple case - with noise - large difference in experiences
sysd = systemd.SystemDynamicsWithSdeint(number_of_agents= 100, k = 0.5, alpha = 2, utility_of_choices= [10, 100, 20],
                          initial_experiences= [200, 2, 50], discount_rate=0.01, sd = 300, rotation_time = 2000, 
                                        flag = False)

time_vector=np.linspace(0, 1000, 100)
sysd.solve(time_vector)
```

Noise helps in making the correct decision eventually

# Less difference in initial experinces Scenario

```python
# Less difference in initial experinces Scenario
sysd = systemd.SystemDynamicsWithSdeint(number_of_agents= 100, k = 0.5, alpha = 2, utility_of_choices= [10, 100, 20],
                          initial_experiences= [100, 2, 50], discount_rate=0.01, sd = 0, rotation_time = 2000, 
                                        flag = False)

time_vector=np.linspace(0, 1000, 100)
sysd.solve(time_vector)
```

agents prefer choice number 2 which neither has high utility nor high initial experience

```python
# even in such scenerios where agents are stuck in the situation like above noise helps them
sysd = systemd.SystemDynamicsWithSdeint(number_of_agents= 100, k = 0.5, alpha = 2, utility_of_choices= [10, 100, 20],
                          initial_experiences= [100, 2, 50], discount_rate=0.01, sd = 400, rotation_time = 2000, 
                                        flag = False)

time_vector=np.linspace(0, 1000, 100)
sysd.solve(time_vector)
```

Thus it can be seen that noise is useful over here
