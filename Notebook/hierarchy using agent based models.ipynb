## Flat Model for Four Choices 

Without the noise agents are stuck in a choice which initially had a higher utility. The environment is dynamic but it can be observed that agents are not able to follow the changes in the environment.

```python
steps = 1000
rotation_step = 100
flag = True
noise_flag = False
d = agentm.Agent_Based_Decision_Model(number_of_agents= 100, k = 1, alpha = 2,
                                utility_of_choices= [25,50,75,100],
                                initial_experiences=[50, 50, 50, 50],
                                discount_rate=1, noise_standard_deviation=0)

d.run(steps,rotation_step, flag, noise_flag, plotting=True)
```

A noisy environment make agents follow the changes in the environment. Lets look at the standard deviation which gives agent the maximum utility

```python
absg = agab.AgentBasedStatGenerator(number_of_agents=100,
                                    k=1,
                                    alpha=2,
                                    utility_of_choices=[25,50,75,100],
                                    initial_experiences=[50, 50, 50, 50],
                                    discount_rate=1,
                                    step_for_sd=5,
                                    end_sd=5000)


steps = 1000
rotation_step = 100

absg.generate_utility_sweep(steps,rotation_step)
```

Somewhere around 1600-1700 utility is maximum. Lets check this 

```python
steps = 1000
rotation_step = 100
flag = True
noise_flag = True
d = agentm.Agent_Based_Decision_Model(number_of_agents= 100, k = 1, alpha = 2,
                                utility_of_choices= [25,50,75,100],
                                initial_experiences=[50, 50, 50, 50],
                                discount_rate=1, noise_standard_deviation=1700)

d.run(steps,rotation_step, flag, noise_flag, plotting=True)
```

### Observe that changes in the environmrent are being followed.... Anything below or above dont allow any useful information to be transfer 

## Hierarchy in information making 

Lets have an hierarchy in decision making..... The rotation step is the same for both nodes. There is noise present 

```python
steps = 1000
rotation_step = [100, 100]
rotation_flag = True
noise_flag = False
d = hierab.Hierarchy(number_of_agents= 100,
                k = 1, alpha = 2,
                utility_of_choices= [25,50,75,100],
                initial_experiences=[50, 50, 50, 50],
                discount_rate=[1,1,1],
                noise_standard_deviation=[0,0,0])

d.run(steps,rotation_step, rotation_flag, noise_flag, plotting=True)
```

The agents shifted from cluster 0 to clustor 1 indicating that there has been a delayed response given the changes in the environment. We wont be nneding too much noise to make the agent follow the changes in the environment

```python
# observe the decisions are changing.... They are not fixed as compared to the flat model without noise

steps = 1000
rotation_step = [100, 100]
rotation_flag = True
noise_flag = False
d = hierab.Hierarchy(number_of_agents= 100,
                k = 1, alpha = 2,
                utility_of_choices= [25,50,75,100],
                initial_experiences=[50, 50, 50, 50],
                discount_rate=[1,1,1],
                noise_standard_deviation=[0,0,0])

d.run(steps,rotation_step, rotation_flag, noise_flag, plotting=True)
```
