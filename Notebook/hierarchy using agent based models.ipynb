## Flat Model for Four Choices 

Without the noise agents are stuck in a choice which initially had a higher utility. The environment is dynamic but it can be observed that agents are not able to follow the changes in the environment.

```python
steps = 1000
rotation_step = 100
flag = True
noise_flag = False
d = agentm.Agent_Based_Decision_Model(number_of_agents= 100, k = 1, alpha = 2,
                                utility_of_choices= [25,50,75,100],
                                initial_experiences=[50, 50, 50, 50],
                                discount_rate=1, noise_standard_deviation=0)

d.run(steps,rotation_step, flag, noise_flag, plotting=True)
```

A noisy environment make agents follow the changes in the environment. Lets look at the standard deviation which gives agent the maximum utility

```python
absg = agab.AgentBasedStatGenerator(number_of_agents=100,
                                    k=1,
                                    alpha=2,
                                    utility_of_choices=[25,50,75,100],
                                    initial_experiences=[50, 50, 50, 50],
                                    discount_rate=1,
                                    step_for_sd=5,
                                    end_sd=5000)


steps = 1000
rotation_step = 100

absg.generate_utility_sweep(steps,rotation_step)
```

Somewhere around 1600-1700 utility is maximum. Lets check this 

```python
steps = 1000
rotation_step = 100
flag = True
noise_flag = True
d = agentm.Agent_Based_Decision_Model(number_of_agents= 100, k = 1, alpha = 2,
                                utility_of_choices= [25,50,75,100],
                                initial_experiences=[50, 50, 50, 50],
                                discount_rate=1, noise_standard_deviation=1700)

d.run(steps,rotation_step, flag, noise_flag, plotting=True)
```

### Observe that changes in the environmrent are being followed.... Anything below or above dont allow any useful information to be transfer 

## Hierarchy in information making 

In a flat model we cannot have two choices changing at a different rate than the other two choices. In such situations one one of the reason is to include hierarchy.

```python
steps = 1000
rotation_step = [400, 400, 400]
rotation_flag = True
noise_flag = False
d = hierab.Hierarchy(number_of_agents= 100,
                    k = 1, alpha = 2,
                    utility_of_choices= [25,50,75,100],
                    initial_experiences=[50, 50, 50, 50],
                    discount_rate=[1,1,1],
                    noise_standard_deviation=[0,0,0],
                    utility_flag = False)

d.run(steps,rotation_step, rotation_flag, noise_flag, plotting=True)
```

Without the noise, agents were able to make the correct choice in a hierachy when the frequences are the same but observe that the decisions are delayed

```python
# what if the frequncy in the lower nodes is different than the upper node

steps = 1000
rotation_step = [400, 200, 200]
rotation_flag = True
noise_flag = False
d = hierab.Hierarchy(number_of_agents= 100,
                    k = 1, alpha = 2,
                    utility_of_choices= [25,50,75,100],
                    initial_experiences=[50, 50, 50, 50],
                    discount_rate=[1,1,1],
                    noise_standard_deviation=[0,0,0],
                    utility_flag = False)

d.run(steps,rotation_step, rotation_flag, noise_flag, plotting=True)
```

Observe a very delayed response at node-1. Within a node agents cant follow changes in the environment

### Introducing noise

All nodes have the same amount of noise... almost same as the flat model 

```python
steps = 1000
rotation_step = [400, 200, 200]
rotation_flag = True
noise_flag = True
d = hierab.Hierarchy(number_of_agents= 100,
                k = 1, alpha = 2,
                utility_of_choices= [25,50,75,100],
                initial_experiences=[50, 50, 50, 50],
                discount_rate=[1,1,1],
                noise_standard_deviation=[1500,1500,1500],
                utility_flag = False)

d.run(steps,rotation_step, rotation_flag, noise_flag, plotting=True)
```

Observe that there is no noise in node-1 but still the agents are following changes in the environment

```python
steps = 1000
rotation_step = [400, 200, 200]
rotation_flag = True
noise_flag = True
d = hierab.Hierarchy(number_of_agents= 100,
                k = 1, alpha = 2,
                utility_of_choices= [25,50,75,100],
                initial_experiences=[50, 50, 50, 50],
                discount_rate=[1,1,1],
                noise_standard_deviation=[0.001,750,750],
                utility_flag = False)

d.run(steps,rotation_step, rotation_flag, noise_flag, plotting=True)
```
